# Visualizing the dataset (s)
## 2t dataset
This dataset contains the temperature (K) averaged on the population

```python
filename = "lcpb_files/2t_pop.zst"

df_2t = pd.read_parquet(filename) # this reads .zst files efficiently
df_2t
```

```python
# converting date_time into all time informations
print('hours of acquisition:',np.unique(df_2t['hour']))
df_2t['year'] = pd.to_datetime(df_2t.index).year
df_2t['month'] = pd.to_datetime(df_2t.index).month
df_2t['day'] = pd.to_datetime(df_2t.index).day
df_2t['hour'] = pd.to_datetime(df_2t.index).hour
df_2t['minute'] = pd.to_datetime(df_2t.index).minute
df_2t['second'] = pd.to_datetime(df_2t.index).second
df_2t['day_of_week'] = pd.to_datetime(df_2t.index).day_name()
df_2t
```

```python
df_2t['date'] = pd.to_datetime(df_2t[['year','month','day']])
df_2t['year_month'] = df_2t['date'].dt.to_period('M').dt.to_timestamp() # creating a column for year_month (useful for grouping)

monthly_avg = df_2t.groupby(['node_id', 'year_month'])['2t'].mean().reset_index()  # averaging data for month and for state

# plot 
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_avg, x='year_month', y='2t', hue='node_id')
# custom legend (under the plot)
plt.legend(
    title='State ID',
    bbox_to_anchor=(0.5, -0.3),   # Adjust vertical position
    loc='upper center',
    ncol=9,                       # 9 columns
    frameon=False
)

plt.title('Monthly Average of 2t (K) per State')
plt.xlabel('Date')
plt.ylabel('Average 2t (K)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Dataframe merging problem!
# We can use VAC to generate data in points that are unknown!!!!!!!!!!!!!!!!!!!
we have different numbers of rows so we have to decide how to approach the problem

First of all, we take out all data referring not to 2022, 2023, 2024

```python
years = [2022, 2023, 2024]

sel_2t = sel_2t[sel_2t['year'].isin(years)]  # Taking out from datasets all data relative to 2021/2025
sel_tp = sel_tp[sel_tp['year'].isin(years)]
sel_solar = sel_solar[sel_solar['year'].isin(years)]
sel_ws10 = sel_ws10[sel_ws10['year'].isin(years)]
sel_ws100 = sel_ws100[sel_ws100['year'].isin(years)]
sel_actuals = sel_actuals[sel_actuals['year'].isin(years)]

print('rows in sel_2t:', np.shape(sel_2t)[0])
print('rows in sel_tp:', np.shape(sel_tp)[0])
print('rows in sel_solar:', np.shape(sel_solar)[0])
print('rows in sel_ws10:', np.shape(sel_ws10)[0])
print('rows in sel_ws100:', np.shape(sel_ws100)[0])
print('rows in sel_actuals:', np.shape(sel_actuals)[0])
```

```python

print(np.unique(sel_2t['hour']))
print(np.unique(sel_tp['hour']))
print(np.unique(sel_solar['hour']))
print(np.unique(sel_ws10['hour']))
print(np.unique(sel_ws100['hour']))
print(np.unique(sel_actuals['hour']))
```

The hourly acquisition of the data is different: 2t, ws10 and ws100 take data every 3h, tp and solar every 6h and actuals every hour. We can attempt to define 2 major timesteps, one from 6 to 18 (day) and the other from 18 to 6 (night). Hopefully no data is missing on any day...



```python
nans = sel_2t.isna().sum()    # checking for NANs in each df
nans = nans + sel_tp.isna().sum()   
nans = nans + sel_solar.isna().sum()
nans = sel_ws10.isna().sum()   
nans = sel_ws100.isna().sum()   
nans = sel_actuals.isna().sum()  
print(sum(nans))
```

At least there are no NA in selected data...

```python
# this function adds a column to a given dataframe
# if the field 'hour' of a row is between 6-18 it is given 'day', otherwise 'night'
def label_day_night(df, time_col='datetime'):
    df = df.copy()

    # move datetime index to column if needed (in our case date_time is the index of pd.df)
    if time_col not in df.columns:
        df[time_col] = df.index
        df.reset_index(drop=True, inplace=True)

    df[time_col] = pd.to_datetime(df[time_col])
    timezone = df[time_col].dt.tz  # keep timezone info if present
    df['hour'] = df[time_col].dt.hour

    # pick day or night
    df['time_window'] = df['hour'].apply(lambda h: 'day' if 6 <= h < 18 else 'night')

    # prepare empty series with matching timezone
    window_start = pd.Series(pd.NaT, index=df.index, dtype=f'datetime64[ns, {timezone}]' if timezone else 'datetime64[ns]')

    # assign day starts
    is_day = df['time_window'] == 'day'
    window_start[is_day] = df.loc[is_day, time_col].dt.floor('D') + pd.Timedelta(hours=6)

    # assign night starts (handle hours < 6 by subtracting a day)
    is_night = ~is_day
    shifted = df.loc[is_night, time_col] - pd.to_timedelta((df.loc[is_night, 'hour'] < 6).astype(int), unit='D')
    window_start[is_night] = shifted.dt.floor('D') + pd.Timedelta(hours=18)

    df['window_start'] = window_start

    return df
```

```python
new_2t = label_day_night(sel_2t)
new_tp = label_day_night(sel_tp)
new_solar = label_day_night(sel_solar)
new_ws10 = label_day_night(sel_ws10)
new_ws100 = label_day_night(sel_ws100)
new_actuals = label_day_night(sel_actuals)

print(np.shape(new_2t)[0])
print(np.shape(new_tp)[0])
print(np.shape(new_solar)[0])
print(np.shape(new_ws10)[0])
print(np.shape(new_ws100)[0])
print(np.shape(new_actuals)[0])
new_2t
```

### Aggregating all data such that we have the same number of rows in each data frame

```python
agg_2t = new_2t.groupby(['node_id','time_window','window_start']).agg(
    t = ('2t','mean')
).reset_index() # aggregating every dataframe, such that node_id, time_window and window_start univoquely
# define an "event"
print('rows in agg_2t:',np.shape(agg_2t)[0])

agg_tp = new_tp.groupby(['node_id','time_window','window_start']).agg(
    tp = ('tp','sum')
).reset_index()
print('rows in agg_tp:',np.shape(agg_tp)[0])

agg_solar = new_solar.groupby(['node_id','time_window','window_start']).agg(
    solar = ('ssrd','sum')
).reset_index()
print('rows in agg_solar:',np.shape(agg_solar)[0])

agg_ws10 = new_ws10.groupby(['node_id','time_window','window_start']).agg(
    ws10 = ('ws_10m','sum')
).reset_index()
print('rows in agg_ws10:',np.shape(agg_ws10)[0])

agg_ws100 = new_ws100.groupby(['node_id','time_window','window_start']).agg(
    ws100 = ('ws_100m','sum')
).reset_index()
print('rows in agg_ws100:',np.shape(agg_ws100)[0])

agg_actuals = new_actuals.groupby(['node_id','time_window','window_start']).agg(
    solar_a = ('solar','sum'),
    wind_a = ('wind','sum'),
    hydro_a = ('hydro','sum'),
    generation_a = ('generation','sum'),
    load_a = ('load','sum')
).reset_index()
print('rows in agg_actuals:',np.shape(agg_actuals)[0])
#print(len(np.unique(agg_actuals['window_start'])))
#print(len(np.unique(agg_tp['window_start'])))
agg_2t
```

### Merging the dataframes

```python
from functools import reduce
import os

dfs = [agg_2t, agg_tp, agg_solar, agg_ws10, agg_ws100, agg_actuals]
df_ITALY = reduce(lambda left, right:
                  pd.merge(left, right,
                    on = ['node_id','time_window','window_start'], # columns used to merge
                    how = 'inner' # exclude not merging lines
                   ), dfs)
df_ITALY

# save this dataframes for future loads
folder = 'output'
filename = 'ITALY_dataset.csv'
fullpath = os.path.join(folder, filename)

# makedir
os.makedirs(folder, exist_ok = True)

# save
df_ITALY.to_csv(fullpath, index = False)
df_ITALY
```
